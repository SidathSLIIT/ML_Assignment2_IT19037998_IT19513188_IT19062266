```python
binary_mapping = {'No': 0, 'Yes': 1}

customer_data['Dependents'] = customer_data['Dependents'].map(binary_mapping)
customer_data['PhoneService'] = customer_data['PhoneService'].map(binary_mapping)
customer_data['Partner'] = customer_data['Partner'].map(binary_mapping)
customer_data['PaperlessBilling'] = customer_data['PaperlessBilling'].map(binary_mapping)
customer_data['Churn'] = customer_data['Churn'].map(binary_mapping)

customer_data
```

```python
internet_service_mapping = {'No': 0, 'Yes': 1, 'No internet service': 0}

customer_data['OnlineBackup'] = customer_data['OnlineBackup'].map(internet_service_mapping)
customer_data['OnlineSecurity'] = customer_data['OnlineSecurity'].map(internet_service_mapping)
customer_data['TechSupport'] = customer_data['TechSupport'].map(internet_service_mapping)
customer_data['DeviceProtection'] = customer_data['DeviceProtection'].map(internet_service_mapping)
customer_data['StreamingMovies'] = customer_data['StreamingMovies'].map(internet_service_mapping)
customer_data['StreamingTV'] = customer_data['StreamingTV'].map(internet_service_mapping)

customer_data
```

```python
phone_service_mapping = {'No': 0, 'Yes': 1, 'No phone service': 0}

customer_data['MultipleLines'] = customer_data['MultipleLines'].map(phone_service_mapping)

customer_data
```

```python
fiber_dsl_mapping = {'Fiber optic': 1, 'DSL': 1, 'No': 0}

customer_data['InternetService'] = customer_data['InternetService'].map(fiber_dsl_mapping)

customer_data
```

```python
contract_mapping = {'Month-to-month': 'MM', 'One year': 'OY', 'Two year': 'TY'}

customer_data['Contract'] = customer_data['Contract'].map(contract_mapping)

customer_data
```

```python
payment_mapping = {'Bank transfer (automatic)': 'Bank', 'Credit card (automatic)': 'Card', 'Electronic check': 'Check','Mailed check': 'Check'}

customer_data['PaymentMethod'] = customer_data['PaymentMethod'].map(payment_mapping)

customer_data
```

```python
# Get the gender as dummy variables

genders = pd.get_dummies(customer_data['gender'], drop_first=True)
genders
```

```python
# Get the contracts as dummy variables

contracts = pd.get_dummies(customer_data['Contract'], drop_first=True)
contracts
```

```python
# Get the payments methods as dummy variables

payments = pd.get_dummies(customer_data['PaymentMethod'], drop_first=True)
payments
```

```python
# View the first five rows in genders column

genders.head()
```

```python
# View the first five rows in payments column

payments.head()
```

```python
# View the first five rows in contracts column

contracts.head()
```

```python
# Concatenate the dummy varibalses to customer_data

customer_data = pd.concat([customer_data,genders,contracts,payments], axis=1)
customer_data
```

```python
# Deleting the unwanted coluns from the customer_data 

customer_data.drop(['customerID','gender','PaymentMethod','Contract'], axis =1, inplace=True)
customer_data
```

```python
# Encoding the dataset to be machine learning friendly

from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()

customer_data_tf = customer_data.copy()

for i in customer_data.columns:
    customer_data_tf[i] = labelEncoder.fit_transform(customer_data[i])

# Encoded dataset
customer_data_tf
```

```python
# Seprating the data set to features and labels 
# Label is Churn coulmn
# Features are other coulmns 
# We get the X axis for Features & Y axis for Lable

X = customer_data_tf.drop('Churn', axis = 1)
Y = customer_data_tf['Churn']
```

```python
# Split the dataset to train(80%) and Test(20%)

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state= 2)
```

```python
# View the shape of created datasets

print(X_train.shape, '---> Shape of training features')
print(X_test.shape, '---> Shape of testing features')
print(Y_train.shape, '---> Shape of training labels')
print(Y_test.shape, '---> Shape of testing labels')
```

```python
#Logistic (sigmoid) Regression 
# We can output the descrete values 
# Sigmoid function has vales of range 0 to 1 
# If it under 0.5 it says its not churned 
# if it over 0.5 it says its churned

from sklearn.linear_model import LogisticRegression

customer_log_model = LogisticRegression(max_iter=150)
customer_log_model.fit(X_train,Y_train)
```

```python
# Getting the churn predictions

customer_predictions = customer_log_model.predict(X_test)
customer_predictions
```

```python
# Check the accuracy of the model using test data set

from sklearn.metrics import classification_report

print(classification_report(Y_test, customer_predictions, target_names=['Retained','Not Retained']))
```

```python
# Getting the accuracy of the model

import sklearn.metrics as sm

customer_model_accuracy = sm.accuracy_score(Y_test,customer_predictions)

print('Accuracy Score ==', str(customer_model_accuracy))
```
